## O(n^2) Sorting Algorithms

Count the number of comparisons (selection vs bubble vs insertion)

* Best case comparison (pre sorted sequences)
* Worst case comparison (reverse sorted sequences)
* Average case comparison (random numbers)

```python
import random
from algorithms import selection_sort, insertion_sort, bubble_sort, shell_sort, 
shell_sort_v1, bucket_sort, radix_sort_counting, radix_sort_insertion, merge_sort, quick_sort

random_collection = []
sorted_collection = []
reverse_collection = []

for size in range(1, 301):
    l = random.sample(range(0, 300), size)
    rand_temp = []
    rev_temp = []
    sort_temp = []
    random_collection.append(l)
    sorted_collection.append(sorted(l))
    reverse_collection.append(sorted(l, reverse=True))
```

```python
from array import array
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 4

random_insertion_comparisons = []
random_bubble_comparisons = []
random_selection_comparisons = []

for arr in random_collection:
    insertion_count = insertion_sort(list(arr))[0]
    bubble_count = selection_sort(list(arr))[0]
    selection_count = bubble_sort(list(arr))[0]
    
    random_insertion_comparisons.append(insertion_count);
    random_bubble_comparisons.append(bubble_count);
    random_selection_comparisons.append(selection_count);
```

```python
sorted_insertion_comparisons = []
sorted_bubble_comparisons = []
sorted_selection_comparisons = []

for arr in sorted_collection:
    insertion_count = insertion_sort(list(arr))[0]
    bubble_count = selection_sort(list(arr))[0]
    selection_count = bubble_sort(list(arr))[0]
    
    sorted_insertion_comparisons.append(insertion_count);
    sorted_bubble_comparisons.append(bubble_count);
    sorted_selection_comparisons.append(selection_count);
```

```python
reverse_insertion_comparisons = []
reverse_bubble_comparisons = []
reverse_selection_comparisons = []

for arr in reverse_collection:
    insertion_count = insertion_sort(list(arr))[0]
    bubble_count = selection_sort(list(arr))[0]
    selection_count = bubble_sort(list(arr))[0]
    
    reverse_insertion_comparisons.append(insertion_count);
    reverse_bubble_comparisons.append(bubble_count);
    reverse_selection_comparisons.append(selection_count);
```

```python
from matplotlib.ticker import NullFormatter

plt.figure(1)
plt.subplot(131)
plt.title('Random Sequences')
plt.plot(random_insertion_comparisons, label='Insertion', linewidth=2)
plt.plot(random_bubble_comparisons, label='Bubble', linewidth=2)
plt.plot(random_selection_comparisons, label='Selection', linewidth=2)
plt.xlabel("Number of Elements")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid(True)

plt.subplot(132)
plt.title('Sorted Sequences')
plt.plot(sorted_insertion_comparisons, label='Insertion', linewidth=2)
plt.plot(sorted_bubble_comparisons, label='Bubble', linewidth=2)
plt.plot(sorted_selection_comparisons, label='Selection', linewidth=2)
plt.xlabel("Number of Elements")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid(True)

plt.subplot(133)
plt.plot(reverse_insertion_comparisons, label='Insertion', linewidth=2)
plt.plot(reverse_bubble_comparisons, label='Bubble', linewidth=2)
plt.plot(reverse_selection_comparisons, label='Selection', linewidth=2)
plt.title('Reverse Sequences')
plt.xlabel("Number of Elements")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid(True)

plt.gca().yaxis.set_minor_formatter(NullFormatter())


plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)
plt.show()
```

## Bucket sort optimization
- Bucket sort on different distributions (use insertion sort as subroutine)

```python
X=np.array(np.loadtxt(open('dataset_random.txt'), dtype=int, delimiter=","))
len(list(X))

distributions = []
```

```python
import matplotlib.mlab as mlab
from random import shuffle

sigma_coefs = [640, 225, 80]

for sigma in sigma_coefs:
    x = 2500+sigma*np.random.randn(10000)
    x = x.astype(int)
    x = list(x)
    x.extend(X)
    shuffle(x)
    distributions.append(x)
```

```python
comparisons = []

for distribution in distributions:
    temp = []
    for i in range(1, 3001):
        temp.append(bucket_sort(list(distribution[(len(distribution)-i):]))[0])
    comparisons.append(temp)
```

```python
rcParams['figure.figsize'] = 15, 15
i = 1
j = 0
for distribution in distributions:
    plt.subplot(3,2,i)
    plt.plot(comparisons[j], linewidth=2)
    plt.axis([0, 3000, 0, 180000])
    plt.xlabel("Number of Elements")
    plt.ylabel("Running Time")
    plt.grid(True)
    i+=1
    plt.subplot(3,2,i)
    plt.hist(distribution, bins = 100, alpha=0.75)
    plt.axis([0, 5000, 0, 2500])
    plt.xlabel("Elements")
    plt.ylabel("Frequency Counts")
    plt.grid(True)
    i+=1
    j+=1
plt.show()
```

## Radix sort optimization
Radix sort with counting sort as subroutine vs radix sort with insertion sort (this is a noncomparison sort, so count the number of element acceses in the sequence)
* pre sorted
* reverse sorted
* Random

```python
random_radix_sort_counting = []
sorted_radix_sort_counting = []
reverse_radix_sort_counting = []
random_radix_sort_insertion = []
sorted_radix_sort_insertion = []
reverse_radix_sort_insertion = []

for size in range(0, 300):
    random_radix_sort_counting.append(radix_sort_counting(list(random_collection[size])))
    sorted_radix_sort_counting.append(radix_sort_counting(list(sorted_collection[size])))
    reverse_radix_sort_counting.append(radix_sort_counting(list(reverse_collection[size])))
    random_radix_sort_insertion.append(radix_sort_insertion(list(random_collection[size])))
    sorted_radix_sort_insertion.append(radix_sort_insertion(list(sorted_collection[size])))
    reverse_radix_sort_insertion.append(radix_sort_insertion(list(reverse_collection[size])))
```

```python
rcParams['figure.figsize'] = 15, 4

from matplotlib.ticker import NullFormatter
plt.figure(1)
plt.subplot(131)
plt.title('Random Sequences')
plt.plot(random_radix_sort_counting, label='Counting', linewidth=2)
plt.plot(random_radix_sort_insertion, label='Insertion', linewidth=2)
plt.xlabel("Number of Elements")
plt.ylabel("Number of Accesses")
plt.legend()
plt.grid(True)

plt.subplot(132)
plt.title('Sorted Sequences')
plt.plot(sorted_radix_sort_counting, label='Counting', linewidth=2)
plt.plot(sorted_radix_sort_insertion, label='Insertion', linewidth=2)
plt.xlabel("Number of Elements")
plt.ylabel("Number of Accesses")
plt.legend()
plt.grid(True)

plt.subplot(133)
plt.plot(reverse_radix_sort_counting, label='Counting', linewidth=2)
plt.plot(reverse_radix_sort_insertion, label='Insertion', linewidth=2)
plt.title('Reverse Sequences')
plt.xlabel("Number of Elements")
plt.ylabel("Number of Accesses")
plt.legend()
plt.grid(True)

plt.gca().yaxis.set_minor_formatter(NullFormatter())


plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)
plt.show()
```
